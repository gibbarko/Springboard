```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
pd.options.display.max_columns = 1000
```

```python
file =  "https://opendata-downloads.s3.amazonaws.com/real_estate_tax_delinquencies.csv"
delins = pd.read_csv(file, index_col=None)
```

```python
zip_counts = pd.DataFrame(delins['zip_code'].value_counts())
zip_counts.head()
```

```python
delins.head()
```

```python
zip_counts.plot(kind='bar')
plt.show()
```

```python
X = delins[['opa_number','zip_code', 'zip_4', 'penalty_due',
               'interest_due', 'other_charges_due', 'total_due', 
               'total_assessment', 'taxable_assessment', 
               'lat', 'lng', 'principal_due']]
```

```python
X = pd.get_dummies(X, columns=['zip_code'])
```

```python
X.head()
```

```python
X = X.dropna()
```

```python
y = X['principal_due']
X = X.drop(columns=['principal_due','total_due'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
reg = LinearRegression().fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
reg.predict(x_train[:3])
```

```python
pd.Series(reg.coef_.round(3), index=x_train.columns).sort_values()
```

```python
y_train[:3]
```

```python
plt.scatter(x=x_train['lng'], y=(y_train - reg.predict(x_train)), color='b')
```
